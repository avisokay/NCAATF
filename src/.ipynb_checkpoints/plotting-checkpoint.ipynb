## Functions

```python
def plot_overlapping_distributions(df, year, division, sex, event, plot_type='kde'):
    """
    Creates overlapping distribution plots for TIME_SECS across different years
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Main dataframe with track and field data
    year : int or list
        Single year or list of years to plot
    division : str
        Division filter (e.g., 'D1')
    sex : str
        Sex filter (e.g., 'Men', 'Women')
    event : str
        Event filter (e.g., '1500', '100')
    plot_type : str
        Type of plot: 'kde', 'hist', or 'both'
    
    Returns:
    --------
    matplotlib.figure.Figure
    """
    
    # Convert single year to list for consistent handling
    if isinstance(year, (int, float)):
        years = [year]
    else:
        years = year
    
    # Filter data for the specified parameters
    filtered_data = df[
        (df['DIVISION'] == division) & 
        (df['SEX'] == sex) & 
        (df['EVENT'] == event) &
        (df['CHAMP_YEAR'].isin(years))
    ]
    
    if filtered_data.empty:
        print(f"No data found for {division} {sex} {event} in years {years}")
        return None
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plot distribution for each year
    for yr in years:
        year_data = filtered_data[filtered_data['CHAMP_YEAR'] == yr]['TIME_SECS']
        
        if not year_data.empty:
            if plot_type in ['kde', 'both']:
                sns.kdeplot(data=year_data, label=f'{yr}', ax=ax)
            
            if plot_type in ['hist', 'both']:
                ax.hist(year_data, alpha=0.3, label=f'{yr}', bins=15, density=True)
    
    # Customize the plot
    ax.set_xlabel('Time (seconds)')
    ax.set_ylabel('Density')
    ax.set_title(f'{division} {sex} {event} - Time Distributions')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # For track events, typically faster times (lower values) are better
    # You might want to invert x-axis depending on the event
    # ax.invert_xaxis()  # Uncomment if needed
    
    plt.tight_layout()
    return fig

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.patches import FancyArrowPatch
from matplotlib.collections import LineCollection
from matplotlib.ticker import FuncFormatter

def plot_yearly_comparison(df, division, sex, event, start_year=None, end_year=None, save_path=None, colormap='plasma'):
    """
    Creates overlapping distributions for all available years within a range
    with small histograms and KDE lines using ordinal color scheme
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Main dataframe
    division : str
        Division filter
    sex : str
        Sex filter  
    event : str
        Event filter
    start_year : int, optional
        Starting year (defaults to earliest in data)
    end_year : int, optional
        Ending year (defaults to latest in data)
    save_path : str, optional
        Path to save the figure as PDF
    colormap : str, optional
        Matplotlib colormap name
    """
    
    # Function to format seconds as MM:SS.s
    def seconds_to_time_format(x, pos):
        """Convert seconds to MM:SS.s format"""
        if x <= 0:
            return "0:00.0"
        
        minutes = int(x // 60)
        seconds = x % 60
        
        if minutes > 0:
            return f"{minutes}:{seconds:04.1f}"
        else:
            return f"{seconds:.1f}s"
    
    # Get available years for the specified parameters
    available_years = df[
        (df['DIVISION'] == division) & 
        (df['SEX'] == sex) & 
        (df['EVENT'] == event)
    ]['CHAMP_YEAR'].unique()
    
    if start_year:
        available_years = available_years[available_years >= start_year]
    if end_year:
        available_years = available_years[available_years <= end_year]
    
    available_years = sorted(available_years)
    
    if len(available_years) == 0:
        print(f"No data found for {division} {sex} {event}")
        return None
    
    # Filter data for all years at once
    filtered_data = df[
        (df['DIVISION'] == division) & 
        (df['SEX'] == sex) & 
        (df['EVENT'] == event) &
        (df['CHAMP_YEAR'].isin(available_years))
    ]
    
    if filtered_data.empty:
        print(f"No data found for {division} {sex} {event} in years {available_years}")
        return None

    # Use the specified colormap
    colors = getattr(plt.cm, colormap)(np.linspace(0.1, 0.9, len(available_years)))
    
    # Set white background style
    plt.style.use('default')
    
    # Create single plot
    fig, ax = plt.subplots(figsize=(14, 8))
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')
    
    # Get overall data range for consistent binning
    all_times = filtered_data['TIME_SECS'].dropna()
    x_min, x_max = all_times.min(), all_times.max()
    x_range = x_max - x_min
    x_min -= x_range * 0.05  # Add some padding
    x_max += x_range * 0.05
    
    # Create consistent bins for all histograms
    bins = np.linspace(x_min, x_max, 25)
    
    # First, get the maximum KDE density to scale histograms
    max_kde_density = 0
    for i, yr in enumerate(available_years):
        year_data = filtered_data[filtered_data['CHAMP_YEAR'] == yr]['TIME_SECS'].dropna()
        if not year_data.empty:
            # Calculate KDE to find max density
            kde_line = sns.kdeplot(data=year_data, ax=plt.figure().add_subplot(), alpha=0)
            kde_density = kde_line.get_lines()[0].get_ydata()
            max_kde_density = max(max_kde_density, kde_density.max())
            plt.close()  # Close the temporary figure
    
    # Plot compressed histograms and KDE for each year
    for i, yr in enumerate(available_years):
        year_data = filtered_data[filtered_data['CHAMP_YEAR'] == yr]['TIME_SECS'].dropna()
        
        if not year_data.empty:
            color = colors[i]
            
            # Calculate histogram data
            hist_counts, hist_bins = np.histogram(year_data, bins=bins, density=True)
            
            # Compress histogram to be only 8% of max KDE density
            compression_factor = 0.08 * max_kde_density / hist_counts.max() if hist_counts.max() > 0 else 1
            compressed_counts = hist_counts * compression_factor
            
            # Plot compressed histogram as bars
            bin_centers = (hist_bins[:-1] + hist_bins[1:]) / 2
            bin_width = hist_bins[1] - hist_bins[0]
            
            ax.bar(bin_centers, compressed_counts, width=bin_width * 0.8, 
                   alpha=0.3, color=color, edgecolor='none')
            
            # KDE line with higher prominence
            sns.kdeplot(data=year_data, ax=ax, color=color, linewidth=4, 
                       alpha=0.9, label=f'{yr}')
    
    # Flip x-axis so shorter times (better performance) are on the right
    ax.invert_xaxis()
    
    # Format x-axis to show MM:SS.s instead of raw seconds
    formatter = FuncFormatter(seconds_to_time_format)
    ax.xaxis.set_major_formatter(formatter)
    
    # ADD COLOR GRADIENT ARROW ANNOTATION
    def add_gradient_arrow(ax, x_start, x_end, y_pos, colors, n_segments=100):
        """Add a smooth gradient arrow showing slow to fast progression"""
        # Create many more segments for smoothness
        x_segments = np.linspace(x_start, x_end, n_segments + 1)
        y_segments = [y_pos] * (n_segments + 1)
        
        # Create line segments
        points = np.array([x_segments, y_segments]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)
        
        # Create smooth color gradient
        segment_colors = getattr(plt.cm, colormap)(np.linspace(0.9, 0.1, n_segments))
        
        # Create line collection with smooth gradient
        lc = LineCollection(segments, colors=segment_colors, linewidth=10, alpha=0.9, capstyle='round')
        ax.add_collection(lc)
        
        # Add arrowhead
        arrow = FancyArrowPatch((x_segments[-5], y_pos), (x_segments[-1], y_pos),
                               arrowstyle='->', mutation_scale=30, 
                               color=segment_colors[-1], linewidth=0)
        ax.add_patch(arrow)
        
        # Add text labels
        ax.text(x_start - (x_end - x_start) * 0.25, y_pos, 'Recent', 
               fontsize=18, ha='right', va='center', fontweight='normal')
        ax.text(x_end + (x_end - x_start) * 0.25, y_pos, '  Older', 
               fontsize=18, ha='left', va='center', fontweight='normal')
        ax.text((x_start + x_end) / 2, y_pos - max_kde_density * 0.05, 'Time Progression', 
               fontsize=18, ha='center', va='top', style='italic')
    
    # Get current y-axis limits to position the arrow
    y_min, y_max = ax.get_ylim()
    
    # Position the gradient arrow in the upper portion of the plot
    arrow_y_pos = y_max * 0.85  # Position at 15% of max y value
    arrow_x_start = x_max - (x_max - x_min) * 0.75  # Start at 75% from right
    arrow_x_end = x_max - (x_max - x_min) * 0.45    # End at 45% from right
    
    add_gradient_arrow(ax, arrow_x_start, arrow_x_end, arrow_y_pos, colors)
    
    # Customize plot with larger fonts
    ax.set_xlabel('Time - Faster â†’', fontsize=20, fontweight='normal')
    ax.set_ylabel('Density', fontsize=20)
    ax.set_title(f'{division} {sex}\'s {event} meters - Distribution of Top Performances Each Year', 
                fontsize=18, fontweight='normal', pad=20)
    
    # Put legend on the figure itself (top right)
    legend = ax.legend(title='Year', loc='upper right', 
                      title_fontsize=12, fontsize=15, frameon=True, 
                      fancybox=False, shadow=False, facecolor='white', 
                      edgecolor='white', framealpha=0.9)
    legend.get_title().set_fontweight('bold')
    
    # White background with prominent gridlines
    ax.grid(True, alpha=0.5, linewidth=0.4, color='lightgray', linestyle='-')
    ax.set_xlim(x_max, x_min)  # Set limits explicitly for inverted axis
    
    # Clean styling with visible borders
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    for spine in ax.spines.values():
        spine.set_linewidth(1)
        spine.set_color('black')
    
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Figure saved to {save_path}")
    
    return fig
```

## Plot Regression plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib.ticker import FuncFormatter

def plot_regression_analysis(df, division, sex, event, save_path=None):
    """
    Creates a regression plot showing performance trends over time with 2021 comparison
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Main dataframe
    division : str
        Division filter
    sex : str
        Sex filter  
    event : str
        Event filter
    save_path : str, optional
        Path to save the figure as PDF
    """
    
    # Function to format seconds as MM:SS.s
    def seconds_to_time_format(x, pos):
        """Convert seconds to MM:SS.s format"""
        if x <= 0:
            return "0:00.0"
        
        minutes = int(x // 60)
        seconds = x % 60
        
        if minutes > 0:
            return f"{minutes}:{seconds:04.1f}"
        else:
            return f"{seconds:.1f}s"
    
    # Filter data for the specified parameters
    filtered_data = df[
        (df['DIVISION'] == division) & 
        (df['SEX'] == sex) & 
        (df['EVENT'] == event)
    ]
    
    if filtered_data.empty:
        print(f"No data found for {division} {sex} {event}")
        return None
    
    # Calculate yearly statistics
    yearly_stats = []
    available_years = sorted(filtered_data['CHAMP_YEAR'].unique())
    
    for year in available_years:
        year_data = filtered_data[filtered_data['CHAMP_YEAR'] == year]['TIME_SECS'].dropna()
        
        if not year_data.empty and len(year_data) > 1:  # Need at least 2 points for std error
            mean_time = year_data.mean()
            std_error = year_data.std() / np.sqrt(len(year_data))
            ci_95 = 1.96 * std_error  # 95% confidence interval
            
            yearly_stats.append({
                'year': year,
                'mean': mean_time,
                'std_error': std_error,
                'ci_95': ci_95,
                'n_observations': len(year_data)
            })
    
    stats_df = pd.DataFrame(yearly_stats)
    
    if stats_df.empty:
        print(f"Insufficient data for regression analysis")
        return None
    
    # Separate pre-2021 and 2021 data
    pre_2021_data = stats_df[stats_df['year'] < 2021].copy()
    data_2021 = stats_df[stats_df['year'] == 2021].copy()
    
    # Set white background style
    plt.style.use('default')
    
    # Create plot
    fig, ax = plt.subplots(figsize=(14, 8))
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')
    
    # Plot pre-2021 data with error bars in blue
    if not pre_2021_data.empty:
        ax.errorbar(pre_2021_data['year'], pre_2021_data['mean'], 
                   yerr=pre_2021_data['ci_95'], 
                   fmt='o', color='steelblue', markersize=8, 
                   capsize=5, capthick=2, elinewidth=2,
                   label='Pre-2021 Annual Means', alpha=0.8)
        
        # Calculate and plot regression line with confidence interval
        if len(pre_2021_data) > 1:
            x_reg = pre_2021_data['year'].values
            y_reg = pre_2021_data['mean'].values
            
            # Perform linear regression
            slope, intercept, r_value, p_value, std_err = stats.linregress(x_reg, y_reg)
            
            # Create extended x range for prediction (including 2021)
            x_extended = np.linspace(x_reg.min(), 2021, 100)
            y_pred = slope * x_extended + intercept
            
            # Calculate prediction interval
            n = len(x_reg)
            x_mean = np.mean(x_reg)
            sxx = np.sum((x_reg - x_mean)**2)
            sxy = np.sum((x_reg - x_mean) * (y_reg - np.mean(y_reg)))
            syy = np.sum((y_reg - np.mean(y_reg))**2)
            s_yx = np.sqrt((syy - sxy**2/sxx) / (n-2))
            
            # 95% confidence interval for the regression line
            t_val = stats.t.ppf(0.975, n-2)
            ci_reg = t_val * s_yx * np.sqrt(1/n + (x_extended - x_mean)**2/sxx)
            
            # Plot regression line and confidence interval
            ax.plot(x_extended, y_pred, color='darkblue', linewidth=3, 
                   label=f'Regression Line (RÂ² = {r_value**2:.3f})')
            ax.fill_between(x_extended, y_pred - ci_reg, y_pred + ci_reg, 
                           color='lightblue', alpha=0.3, 
                           label='95% Confidence Interval')
            
            # Calculate predicted 2021 value
            pred_2021 = slope * 2021 + intercept
            ci_2021 = t_val * s_yx * np.sqrt(1/n + (2021 - x_mean)**2/sxx)
    
    # Plot 2021 data in red if available
    if not data_2021.empty:
        ax.errorbar(data_2021['year'], data_2021['mean'], 
                   yerr=data_2021['ci_95'], 
                   fmt='o', color='red', markersize=10, 
                   capsize=6, capthick=3, elinewidth=3,
                   label='2021 Observed Mean', alpha=0.9, zorder=5)
        
        # Add comparison text if we have regression
        if not pre_2021_data.empty and len(pre_2021_data) > 1:
            observed_2021 = data_2021['mean'].iloc[0]
            difference = observed_2021 - pred_2021
            
            # Add annotation showing the difference
            ax.annotate(f'Difference from predicted:\n{difference:+.2f} seconds', 
                       xy=(2021, observed_2021), xytext=(2021-2, observed_2021),
                       fontsize=12, ha='right', va='center',
                       bbox=dict(boxstyle="round,pad=0.3", facecolor='yellow', alpha=0.7),
                       arrowprops=dict(arrowstyle='->', color='red', lw=2))
    
    # Format y-axis to show MM:SS.s instead of raw seconds
    formatter = FuncFormatter(seconds_to_time_format)
    ax.yaxis.set_major_formatter(formatter)
    
    # DON'T invert y-axis - let times naturally decrease downward over time
    # (Removed: ax.invert_yaxis())
    
    # Customize plot
    ax.set_xlabel('Year', fontsize=20, fontweight='normal')
    ax.set_ylabel('Mean Performance Time', fontsize=20)
    ax.set_title(f'{division} {sex}\'s {event} meters - Performance Trends Over Time', 
                fontsize=18, fontweight='normal', pad=20)
    
    # Legend
    legend = ax.legend(loc='best', fontsize=12, frameon=True, 
                      fancybox=False, shadow=False, facecolor='white', 
                      edgecolor='gray', framealpha=0.9)
    
    # Grid and styling
    ax.grid(True, alpha=0.5, linewidth=0.4, color='lightgray', linestyle='-')
    
    # Clean styling with visible borders
    for spine in ax.spines.values():
        spine.set_linewidth(1)
        spine.set_color('black')
    
    # Set reasonable x-axis limits
    if not stats_df.empty:
        x_range = stats_df['year'].max() - stats_df['year'].min()
        ax.set_xlim(stats_df['year'].min() - x_range*0.05, 
                   stats_df['year'].max() + x_range*0.05)
    
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"Figure saved to {save_path}")
    
    # Print summary statistics
    if not pre_2021_data.empty and len(pre_2021_data) > 1:
        print(f"\nRegression Analysis Summary:")
        print(f"Slope: {slope:.4f} seconds/year")
        print(f"R-squared: {r_value**2:.3f}")
        print(f"P-value: {p_value:.4f}")
        if not data_2021.empty:
            print(f"2021 Predicted: {seconds_to_time_format(pred_2021, None)}")
            print(f"2021 Observed: {seconds_to_time_format(observed_2021, None)}")
            print(f"Difference: {difference:+.2f} seconds")
    
    return fig, stats_df
```

```python
# All years in range with save functionality
division = 'D1'
sex = 'Men'
event = '10000'

fig = plot_regression_analysis(df, division, sex, event,
                           save_path=f"../plots/{division}_{sex}_{event}_regression.pdf")
```
